```python
all_skills=[]
```

```python
for skills in df['required_skills']:
    skills_list=skills.split(',')
    for skills in skills_list:
        all_skills.append(skills.strip())
```

```python
for skill in df['preferred_skills']:
    skills_lst=skill.split(',')
    for skill in skills_lst:
        all_skills.append(skill.strip())
```

```python
unique_skills=sorted(set(all_skills))
```

```python
print("All Unique Skills are: ",unique_skills)
```

```python
print("Total Unique Skills: ",len(unique_skills))
```

```python
skills_df=pd.DataFrame({
    'skill_id': range(1,len(unique_skills)+1),
    'skill_name' : unique_skills
})
```
