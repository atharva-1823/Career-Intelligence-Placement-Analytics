```python
import numpy as np
import pandas as pd
```

```python
jobs_df = pd.read_csv("data/jobs.csv")
```

```python
skills_df=pd.read_csv("data/skills_combined.csv")
```

```python
jobs_df = jobs_df.drop_duplicates()
```

```python
jobs_df = jobs_df.dropna(how='all')
```

```python
jobs_df = jobs_df[jobs_df["job_id"] != "job_id"]
```

```python
jobs_df["job_id"].nunique()
```

```python
jobs_df.to_csv("data/jobs.csv",index=False)
```

```python
jobs_df.shape
```

```python
skill_lookup=dict(
    zip(skills_df['skill_name'],skills_df['skill_id'])
)
```

```python
job_skills=[]
```

```python
for _,row in jobs_df.iterrows():
    job_id=row['job_id']

    required_skills= row['required_skills'].split(',')

    for skill in required_skills:
        skill=skill.strip()

        if skill in skill_lookup:
            job_skills.append([
                job_id,
                skill_lookup[skill],
                "Required"
            ])
```

```python
for _,row in jobs_df.iterrows():
    job_id=row['job_id']

    preferred_skills= row["preferred_skills"].split(",")
    for skill in preferred_skills:
        skill=skill.strip()

        if skill in skill_lookup:
            job_skills.append([
                job_id,
                skill_lookup[skill],
                "Preferred"
            ])
```

```python
job_skills_df = pd.DataFrame(
    job_skills,
    columns=[
        "job_id",
        "skill_id",
        "skill_type"
    ]
)
```
